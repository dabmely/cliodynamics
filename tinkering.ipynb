```python
import seaborn as sns
import matplotlib.pyplot as plt

from collections import deque
import logging
import random
import time

from bokeh.layouts import row, layout
from bokeh.plotting import figure, show
from bokeh.palettes import Category20b, Magma
from bokeh.transform import factor_cmap
from bokeh.io import curdoc
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, Slider
from bokeh.models import LinearColorMapper

from cliodynamics.models.frontier import MetaethnicFrontierModel
from cliodynamics.models.frontier import plot_model
```

```python
model = MetaethnicFrontierModel.empty_model(101)

palette = sns.color_palette("Set2", n_colors=num_empires)
random.shuffle(palette)
palette.insert(0, (0.8, 0.8, 0.8))
```

```python
for _ in range(10):
    powers, distances, areas_dense, average_asabiyas_dense, collapsed = model.step()
```

```python
fig, ax = plt.subplots(3, 3)
fig.set_figheight(12)
fig.set_figwidth(12)
fig.set_dpi(100)

ax[0, 0].matshow(powers)
# ax[0, 0].colorbar()
ax[0, 0].set_title("powers")

ax[0, 1].matshow(distances)
# ax[0, 1].colorbar()
ax[0, 1].set_title("distances")

ax[0, 2].matshow(areas_dense)
# ax[0, 2].colorbar()
ax[0, 2].set_title("areas_dense")

ax[1, 0].matshow(average_asabiyas_dense)
# ax[1, 0].colorbar()
ax[1, 0].set_title("average_asabiyas_dense")

ax[1, 1].matshow(collapsed)
# ax[1, 1].colorbar()
ax[1, 1].set_title("collapsed")

ax[1, 2].matshow(model.asabiya)
# ax[1, 2].colorbar()
ax[1, 2].set_title("model:asabiya")

num_empires = 30
sns.heatmap(
    model.membership,
    cmap=palette,
    ax=ax[2, 0],
    xticklabels=False,
    yticklabels=False,
    vmin=0,
    vmax=num_empires,
)
ax[2, 0].set_title("model:membership")
```
